```python
import pandas as pd
import numpy as np
from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
SalaryData_Train = pd.read_csv('C:/Users/abc/Desktop/Assignments/Assignment 14 SVM/SalaryData_Train(1).csv')
SalaryData_Test = pd.read_csv('C:/Users/abc/Desktop/Assignments/Assignment 14 SVM/SalaryData_Test(1).csv')
```

```python
SalaryData_Train.head()
```

```python
#SalaryData_Train.columns
```

```python
SalaryData_Train1 = SalaryData_Train.drop('educationno',axis=1)
SalaryData_Train1
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
# Converting catagorical into label
convert_col=['workclass','education','maritalstatus','occupation','relationship','race','sex','native']
```

```python
for i in convert_col:
    SalaryData_Train1[i]=labelencoder.fit_transform(SalaryData_Train1[i])
SalaryData_Train1
```

```python
#Dividing dataset dataset
X = SalaryData_Train1.iloc[: ,0:12]
Y = SalaryData_Train1.iloc[: , 12]
```

For Hyperparameter using Grid Search CV best fit value for model

```python
clf = SVC()
#param_grid = [{'kernel':['poly'],'gamma':[100,5,15,0.5],'C':[19,14,13,12,11,10,0.1,0.001] }]
#param_grid = [{'kernel':['rbf','poly'],'gamma':[100,5,15,0.5],'C':[12,11,10,0.1,0.001] }]
param_grid = [{'kernel':['rbf'],'gamma':[100,5,15,0.5],'C':[12,11,10,0.1,0.001] }]
gsv = GridSearchCV(clf,param_grid,cv=10)
gsv.fit(X,Y)
```

```python
gsv.best_params_ , gsv.best_score_ 
```

```python
# Using best value findout from gridsearch building model
clf = SVC(C= 12, gamma = 100)
clf.fit(X,Y)
```

```python
SalaryData_Test1 = SalaryData_Test.drop('educationno',axis=1)
SalaryData_Test1 
```

```python
# Converting catagorical into label
convert1_col=['workclass','education','maritalstatus','occupation','relationship','race','sex','native']
```

```python
for i in convert1_col:
    SalaryData_Test1[i]=labelencoder.fit_transform(SalaryData_Test1[i])
SalaryData_Test1
```

```python
#Dividing dataset dataset
X_Test = SalaryData_Test1.iloc[: ,0:12]
Y_Test= SalaryData_Test1.iloc[: , 12]
```

```python
# Calculating Accuracy
clf = SVC(C= 12, gamma = 100)
clf.fit(X , Y)
y_pred = clf.predict(X_Test)
acc = accuracy_score(Y_Test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(Y_Test, y_pred)
```

```python
# For F1 score
print(classification_report(Y_Test,y_pred))
```
